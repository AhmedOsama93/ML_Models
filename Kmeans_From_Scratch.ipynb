```python
#AhmedOsamaMamoud       #20190015
import pandas as pd
import numpy as np
import math
import random
from sklearn.preprocessing import MinMaxScaler
#"D:\\4th year 2nd term\\2DataMining\\Assignments\\Assignment2\\Assignment2\\crime_data.csv"

#preprocessing data
def readData(link):
    data = pd.read_csv(link)
    return data
def normalize(data):
    scaler = MinMaxScaler()
    cols = ['Murder', 'Assault', 'UrbanPopulation', 'Rape']
    data[cols] = scaler.fit_transform(data[cols])
    return data
```

```python

def euclidean_distance(point1, point2):
    distance = 0.0
    for i in range(len(point1)):
        distance += (point1[i] - point2[i]) ** 2
    return distance ** 0.5
def manhattan_distance(point1, point2):
    distance = 0
    for i in range(len(point1)):
        distance += abs(point1[i] - point2[i])
    return distance
def distance(choose, data, centroid):
    distances=[]
    cols = ['Murder', 'Assault', 'UrbanPopulation', 'Rape']
    if choose == 1:
        for ind in range (len(data)):
            distances.append(manhattan_distance(data.loc[ind,cols] ,centroid))
    if choose==2 :
        for ind in range (len(data)):
            distances.append(euclidean_distance(data.loc[ind,cols],centroid))
    return distances
```

```python
def centroidDistanceStTime(choose, numOfCentroid, data):
    df = pd.DataFrame({})
    cols = ['Murder', 'Assault', 'UrbanPopulation', 'Rape']
    for i in range(numOfCentroid):
        centroid= random.randint(1,49)
        distances=distance(choose, data, data.loc[centroid,cols])
        c= 'cluster#'+str(i+1)
        df[c] = np.array(distances)
    return df
def printOutliers(data):
    Q1 = data['Distance'].quantile(0.25)
    Q3 = data['Distance'].quantile(0.75)
    IQR = Q3 - Q1
    res = data[(data['Distance'] < (Q1 - 1.5 * IQR)) | (data['Distance'] > (Q3 + 1.5 * IQR))]
    return res

def centroidDistance(choose,numOfCentroid,data,centroids):
    df = pd.DataFrame({})
    for i in range(numOfCentroid):
        distances=distance(choose, data, centroids[i])
        c= 'cluster#'+str(i+1)
        df[c] = np.array(distances)
    return df

def newCentroids(data, numOfCentroid):
    centroids=[]
    for i in range(numOfCentroid):
        c = 'cluster#' + str(i+1)
        df = data.loc[data['Cluster'] == c]
        avg=[]
        avg.append(df['Murder'].mean())
        avg.append(df['Assault'].mean())
        avg.append(df['UrbanPopulation'].mean())
        avg.append(df['Rape'].mean())
        centroids.append(avg)
    return centroids
```

```python
def Kmeans(data, numOfCentroid, orgData) :
    for i in range(len(data)):
        mn=2
        mnPos=1
        for j in range(numOfCentroid):
            if(data.iloc[i, j]<mn):
                mnPos=j
                mn=data.iloc[i, j]
        orgData.loc[i,['Cluster']]= 'cluster#'+str(mnPos+1)
        orgData.loc[i, ['Distance']] = mn
    return orgData
```
